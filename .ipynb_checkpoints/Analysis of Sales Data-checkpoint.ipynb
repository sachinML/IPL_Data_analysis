```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Merging all the 12 months sales data into a single data

files =[file for file in os.listdir("Sales_Data")]

path = "C:\\Users\\Sachin Jha\\data_analysis_project\\Sales_Data"

#blank dataframe
all_data = pd.DataFrame()

for file in files:
    current_df = pd.read_csv(path+"/"+file)
    all_data = pd.concat([all_data, current_df])

all_data.shape
```

```python
#dataset conversion to csv file 
all_data.to_csv('C:\\Users\\Sachin Jha\\data_analysis_project\\Sales_Data/all_data.csv',index=False)
```

```python
all_data.head()
```

# 
so after observing this dataset we found that there are some nan values so we need to figure it out and remove or replace it

```python
all_data.isnull().sum()  # gives how many nan values are there in each column
```

```python
all_data = all_data.dropna(how='all') #it will drop the rows where all elements are missing(nan values)
all_data.head()
```

# Best month for the sale?

```python
# we will go through the order date  column from the datarame and extract the months from there
def month(x):
    return x.split('/')[0]

all_data['Month'] = all_data['Order Date'].apply(month) # add a new column month in dataframe
pd.DataFrame(all_data['Month']).head()
```

```python
all_data.dtypes # now we can see that order date is also a object type so we need to convert it into integer type
```

```python
# first we will check is there any string data in between month because it will create an error if we directly convert to int type

all_data['Month'].unique()
```

```python
filter = (all_data['Month'] == 'Order Date') # for filtering string type data
all_data = all_data[~filter]

all_data['Month']=all_data['Month'].astype(int) # so month changes to int type now

all_data.dtypes 
```

```python
# now we can observe that price is given in the obeject type too so we need to convert it into a float value

# and also convert the Quantity Ordered in int type

all_data['Price Each'] = all_data['Price Each'].astype(float) 
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
```

```python
all_data.dtypes 
```

```python
# now for finding the best month for the sale we need to find how much sale has happend in each month

all_data['sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

```python
x = all_data.groupby(['Month']).sum()
x
```

```python
# it will return the total sale in each month
y = x.drop(['Quantity Ordered','Price Each'], axis = 1)
y                                                      
```

```python
months = range(1,13)
plt.bar(months, y.sales)
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

# 
so from this we can conclude that December(Month number 12) was best for sale.

# What product sold the most? and what was the reason for that?


```python
all_data.head()
```

```python
#shows the each product in the quantity ordered
products = all_data.groupby('Product')['Quantity Ordered'].sum().index


quantity = all_data.groupby('Product')['Quantity Ordered'].sum()

# gives avegrage price of each product ordered
prices = all_data.groupby('Product')['Price Each'].mean()
```

```python
# we will basically see the sold product on the basis of quantity and price together for better analysis 

plt.figure(figsize=(40,24))
fig,ax1 = plt.subplots()

ax2=ax1.twinx()
ax1.bar(products, quantity, color='b')
ax2.plot(products, prices, 'r-')

ax1.set_xticklabels(products, rotation='vertical', size=8);
```

# 
So we can see that the top selling product is 'AAA Batteries'. 
Now because top selling products seem to have a correlation with the price of the product. 
So we can conclude that the cheaper the product higher the quantity ordered and vice versa.

# Now out of these sold products how much are most often sold together?

```python
all_data.head()
```

```python
# for finding the together sold product we will track the same Order ID

all_data['Order ID'].duplicated(keep=False) # will give the False if not repeated otherwise True
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)] # will return the id which are repeated
df.head(10) 
```

```python
df.shape
```

```python
# it will create a new column name Grouped which will show the sold together item on same Order ID
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
#lets drop out all duplicate Order ID
df2 = df.drop_duplicates(subset=['Order ID'])
df2['Grouped'].value_counts()[0:5].plot.pie()
```

# Which city has max order?

```python
all_data.head()
```

```python
def city(x):
    return x.split(',')[1]

all_data['city'] = all_data['Purchase Address'].apply(city) # add a city column in add_data by applying above method
all_data.head()
```

```python
bs = all_data.groupby('city')['city'].count()
plt.bar(bs.index, bs)
plt.xticks(rotation='vertical')
plt.ylabel('received orders')
plt.xlabel('city names')
plt.show()
```

# 
So we can say that San Francisco has maximum order for sale.
